```python
# 1.Load and prepare a dataset for binary classification.

import pandas as pd
```

```python
data = pd.read_csv('breast-cancer.csv')
data.head()
```

```python
from sklearn.datasets import load_breast_cancer
df = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
df.head()
```

```python
#2.Train an SVM with linear and RBF kernel.
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
x_test, x_train , y_test, y_train  = train_test_split(df.drop('target' , axis=1), df['target'], test_size=0.2, random_state=42)

model = SVC(kernel='linear')
model.fit(x_train , y_train)
y_pred  = model.predict(x_test)
accuracy_score(y_test , y_pred)
```

```python
model.score(x_test,y_pred)
```

```python
model_rbf = SVC(kernel='rbf')
model_rbf.fit(x_train , y_train)
y_pred_rbf  = model_rbf.predict(x_test)
accuracy_score(y_test , y_pred_rbf)
```

```python
model_rbf.score(x_test , y_pred_rbf)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# Select ONLY 2 features
X = df[["mean radius", "mean texture"]].values
y = df["target"].values

# Train SVM ONLY on 2 features
model2 = SVC(kernel='linear')
model2.fit(X, y)

# Create mesh grid
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 400),
    np.linspace(y_min, y_max, 400)
)

# Predict for each point in the grid (now using model2)
Z = model2.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundary
plt.contourf(xx, yy, Z, cmap='coolwarm', alpha=0.3)

# Plot actual data points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolor="black")

plt.xlabel("Mean Radius")
plt.ylabel("Mean Texture")
plt.title("Decision Boundary (SVM Linear Kernel)")
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': ['scale',0.1, 0.01, 0.001],
    'kernel': ['rbf', 'linear']
}

grid = GridSearchCV(SVC(),param_grid,cv=5)
grid.fit(x_train,y_train)

# Best parameters
print("Best Parameters:", grid.best_params_)

# Best model accuracy
print("Best Score:", grid.best_score_)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(grid, x_train, y_train, cv=5)
print("Cross-validation scores:", scores)
print("Mean accuracy:", scores.mean())
```
